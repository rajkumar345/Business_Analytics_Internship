# Cell 2: Import Libraries (Code Cell)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Display all columns when viewing DataFrame
pd.set_option('display.max_columns', None)
```

# Cell 3: Load Dataset (Code Cell)

```python
# Load the dataset from data folder
df = pd.read_csv("../data/Data_set 2.csv")

# Show basic info
print("Shape (rows, columns):", df.shape)
df.head()
```

# Cell 4: Dataset Info (Code Cell)

```python
# Column details and data types
df.info()
```

# Cell 5: Descriptive Stats (Code Cell)

```python
# Summary for numeric and categorical columns
df.describe(include="all")
```

# Cell 6: Missing Values & Duplicates (Code Cell)

```python
print("Missing Values:")
print(df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

# Cell 7: Gender Distribution (Code Cell)

```python
# Clean and analyze Gender column
df['gender'] = df['gender'].astype(str).str.strip().str.title()

# Map similar values
mapping = {
    'M': 'Male', 'Male': 'Male', 'male': 'Male',
    'F': 'Female', 'Female': 'Female', 'female': 'Female',
    'Other': 'Other', 'O': 'Other', 'Non-Binary': 'Other', 'Na': 'Unknown'
}
df['Gender_clean'] = df['gender'].map(mapping).fillna('Unknown')

# Counts and percentages
print(df['Gender_clean'].value_counts())
print("\nPercentages:")
print((df['Gender_clean'].value_counts(normalize=True) * 100).round(2))
```

# Cell 8: Visualization (Code Cell)

```python
# Bar Chart
plt.figure(figsize=(6,4))
df['Gender_clean'].value_counts().plot(kind='bar', color='lightblue')
plt.title("Gender Distribution - Counts")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(6,6))
df['Gender_clean'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.ylabel('')
plt.title("Gender Distribution - Percentages")
plt.tight_layout()
plt.show()
```

# Cell 9: Save Cleaned Data (Code Cell)

```python
df.to_csv("../outputs/cleaned_dataset.csv", index=False)
print("Cleaned dataset saved to outputs/cleaned_dataset.csv")
```
